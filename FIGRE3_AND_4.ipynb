```python
#PSNR SCALE 4 SET14
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib.pyplot import *

psnr_scale_4 = [       27.43,        27.37,     27.24,     27.34,    28.02,     28.35,     27.52,     27.61]
no_of_parameters = [   50,            25,       10,        8,        5,       0.09 ,         0.05,      0.03]
Name_Of_Models = [     "SelfExSR",   "SCN",    "RFL",     "A+" ,     "VDSR",    "LSIR",    "SRCNN",   "FSRCNN"]
for k in range(8):
    loc_x = no_of_parameters[k]
    loc_y = psnr_scale_4[k]
    if loc_x == 0.09 and loc_y == 28.35:
        plt.semilogx(loc_x,loc_y,color='b',marker='.')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=12,color='b')
    else:
        plt.semilogx(loc_x,loc_y,color='r',marker='.')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=12)
    
plt.ylim([27,28.5])
plt.xlim(1000,0.00018)
#plt.xlim(10,10000)
plt.ylabel('PSNR(dB)',fontsize=16)
plt.xlabel('slower  <--------       running time(s)        ---------->        faster',fontsize=12)
plt.grid(which="both")
plt.show()
```
